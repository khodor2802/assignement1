```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python

# Load the dataset
data = pd.read_csv('larger_sales_dataset.csv')

# Total length of the dataset
total_length = len(data)
print("Total length of the dataset:", total_length)
```

```python


# Convertir la colonne 'Order Date' en format de date
data['Order Date'] = pd.to_datetime(data['Order Date'])

# Extraire le mois à partir de la colonne 'Order Date'
data['Month'] = data['Order Date'].dt.month

# Calculer la quantité totale par mois
quantity_by_month = data.groupby('Month')['Quantity'].sum()

# Visualiser la distribution de la quantité par mois
plt.figure(figsize=(10, 6))
quantity_by_month.plot(kind='bar', color='skyblue')
plt.xlabel('Mois')
plt.ylabel('Quantité')
plt.title('Quantité par Mois')
plt.xticks(rotation=0) 
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(data['Quantity'], data['Unit Price'], alpha=0.5)
plt.title('Corrélation entre la Quantité et le Prix Unitaire')
plt.xlabel('Quantité')
plt.ylabel('Prix Unitaire')
plt.grid(True)

mean_quantity = data['Quantity'].mean()
mean_unit_price = data['Unit Price'].mean()
covariance = ((data['Quantity'] - mean_quantity) * (data['Unit Price'] - mean_unit_price)).sum() / len(data)
std_quantity = data['Quantity'].std()
std_unit_price = data['Unit Price'].std()

correlation_coefficient = covariance / (std_quantity * std_unit_price)
print("Correlation Coefficient:", correlation_coefficient)


slope = covariance / std_quantity ** 2
intercept = mean_unit_price - slope * mean_quantity


plt.plot(data['Quantity'], slope * data['Quantity'] + intercept, color='red')

plt.show()
```
